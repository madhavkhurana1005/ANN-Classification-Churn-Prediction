```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

```python
leg = LabelEncoder()
# df['Geography'] = leg.fit_transform(df['Geography'])
df['Gender'] = leg.fit_transform(df['Gender'])
```

```python
df.head(3)
```

```python
# One hot encoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
data_geo =ohe.fit_transform(df[['Geography']])
```

```python
df.head()
```

```python
cols = ohe.get_feature_names_out()
```

```python
pd.DataFrame(data_geo.toarray(),columns=cols).head()
```

```python
data = pd.concat([df.drop(['Geography'],axis=1),pd.DataFrame(data_geo.toarray(),columns=cols)],axis=1)
```

```python
data
```

```python
# Saving in pickle files

with open('leg_data_gender.pkl', 'wb') as f:
    pickle.dump(leg, f)

with open('ohe_data_geo.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
data.head()
```

```python
data.columns.to_list()
```

```python
x = data.drop(['Exited'],axis=1)
y = data['Exited']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

## ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
from datetime import datetime
```

```python
import sys
sys.executable
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
x_train.shape
```

```python
model = Sequential(
   [ Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')]
)
```

```python
model.summary()
```

```python
opt =  tf.keras.optimizers.Adam(learning_rate=0.001 )
```

```python
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Set up the Tensorboard
```

```python
log_dir = "logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# Setting up EarlyStopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Training the model
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs = 50, callbacks=[tensorboard_callback, early_stopping_callback] )
```

```python
model.save("model.h5")
```
